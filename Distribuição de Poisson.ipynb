# Importando Bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn

from scipy.stats import poisson

import warnings
warnings.filterwarnings("ignore")

def drop_reset_index(df):
  df = df.dropna()
  df = df.reset_index(drop=True)
  df.index += 1
  return df
```

# Importando DataBase


```python
df = pd.read_csv("https://www.football-data.co.uk/mmz4281/2223/E0.csv")
df = df[['HomeTeam','AwayTeam','FTHG','FTAG']]
df.columns = ['Home','Away','Goals_H','Goals_A']
df = drop_reset_index(df)
display(df)
```

# DistribuiÃ§Ã£o de Poisson


```python
import plotly.graph_objects as go

pred_goals = [[poisson.pmf(i, df.mean()[j]) for i in range(df['Goals_H'].max()+3)] for j in range(2)]
home_pred = [i*100 for i in pred_goals[0]]
away_pred = [i*100 for i in pred_goals[1]]

goals = ['0','1','2','3','4','5','6','7','8']

fig = go.Figure()
fig.add_trace(go.Histogram(
    x = df['Goals_H'],
    histnorm='percent',
    name='Home',
    marker_color='#FF6F61',
    opacity=0.65
))

fig.add_trace(go.Histogram(
    x = df['Goals_A'],
    histnorm='percent',
    name='Away',
    marker_color='#77DD77',
    opacity=0.65
))

fig.add_trace(go.Scatter(x=goals, y=home_pred,
                         mode='lines+markers',
                         line=dict(color='#FF6F61', width=4, dash='dash'),
                         name='Home Poisson'))

fig.add_trace(go.Scatter(x=goals, y=away_pred,
                         mode='lines+markers',
                         line=dict(color='#77DD77', width=4, dash='dash'),
                         name='Away Poisson'))

fig.update_layout(
    title='NÃºmero de Gols por Partida x DistribuiÃ§Ã£o de Poisson (Premier League 2022/23)',
    xaxis_title='Goals',
    yaxis_title='%',
    barmode='group',
    template='plotly_dark',
    title_font_size=24,
    legend=dict(x=0.01, y=0.99)
)

fig.show()
```

```python
Media_Goals_H = df.mean()[0]
Media_Goals_A = df.mean()[1]
print(f"A mÃ©dia de gols do Mandante Ã© {round(Media_Goals_H,2)} e a mÃ©dia de gols do Visitante Ã© {round(Media_Goals_A,2)}.")

Max_Goals_H = df[['Goals_H','Goals_A']].max()[0]
Max_Goals_A = df[['Goals_H','Goals_A']].max()[1]
print(f"O mÃ¡ximo de gols que o mandante marcou foi {Max_Goals_H} e mÃ¡ximo de gols que o Visitante marcou foi {Max_Goals_A}.")
```
